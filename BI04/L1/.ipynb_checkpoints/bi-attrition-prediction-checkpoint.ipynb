```python
#2. 标签编码

attr=['Age','BusinessTravel','Department','Education','EducationField','Gender','JobRole','MaritalStatus','OverTime']
```

```python
train_df['Attrition'] = train_df['Attrition'].map(lambda x:1 if x=="Yes" else 0)

for feature in attr:
    lbe = LabelEncoder()
    train_df[feature] = lbe.fit_transform(train_df[feature])
    test_df[feature] = lbe.transform(test_df[feature])
#train_df.to_csv('datasets/train_a.csv')
#test_df.to_csv('datasets/test_a.csv')
```
